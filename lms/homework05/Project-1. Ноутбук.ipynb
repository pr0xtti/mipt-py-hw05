Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
import re

# 1. Признак Город
million_cities = [
    'Новосибирск', 
    'Екатеринбург',
    'Нижний Новгород',
    'Казань', 
    'Челябинск',
    'Омск', 
    'Самара', 
    'Ростов-на-Дону', 
    'Уфа', 
    'Красноярск', 
    'Пермь', 
    'Воронеж',
    'Волгоград',
]
city_map: dict = {
    k: "город-миллионник" for k in million_cities
}
city_map["Москва"]          = "Москва"
city_map["Санкт-Петербург"] = "Санкт-Петербург"
src_col = 'Город, переезд, командировки'

df['Город'] = df[src_col].str.split(',', n=1).str[0].str.strip()
df['Город'] = df['Город'].map(lambda x: city_map.get(x, "другие"))
print(df['Город'].unique())

# 2. Признаки: Готовность к переезду, Готовность к командировкам
move_col = 'Готовность к переезду'
trip_col = 'Готовность к командировкам'

def get_values(text: str) -> tuple[bool | None]:
    move_value, trip_value = None, None
    
    move, trip = text.split(',', maxsplit=2)[1:3]
    move, trip = move.lstrip(), trip.lstrip()
    
    if move.startswith(('готов', 'хочу пе')):
        move_value = True
    elif move.startswith('не го'):
        move_value = False
    
    # Найдено значение для переезда => с форматом ОК
    if move_value is not None:  
        if trip.startswith('готов'):
            trip_value = True
        elif trip.startswith('не го'):
            trip_value = False
        
        if trip_value is not None:  # Все найдено
            return move_value, trip_value
            
    # Проблема с форматом, нужно копать
    for item in text.split(',', maxsplit=3)[1:4]:
        item = item.lstrip()
        if move_value is None:
            if item.startswith(
                ('готов к п', 'готова к п', 'хочу п')
            ):
                move_value = True
            elif item.startswith(('не готов к п', 'не готова к п')): 
                move_value = False
            # else:
            #     raise AssertionError(f"Parsing problem: {text}")
        else:
            if item.startswith(
                'не готов к к'
            ):
                trip_value = False
            elif re.match(r'^готов(а)? к (?:редким )?к', item):
                trip_value = True
            
            if (move_value is not None) and (trip_value is not None):
                return move_value, trip_value
                
    #raise AssertionError(f"Parsing problem: {text}")
    return move_value, trip_value

#df = df_org.copy()
if src_col in df.columns:
    tmp = df[src_col].str.replace(r'\([^)]*\)', '', regex=True).map(get_values)
    tmp = pd.DataFrame(tmp)
    df[[move_col, trip_col]] = tmp[src_col].apply(pd.Series)
    df[move_col] = df[move_col].astype('boolean')
    df[trip_col] = df[trip_col].astype('boolean')

    df.drop([src_col], axis=1, inplace=True)
else:
    print(f"Column {src_col} already parsed")

print('\n', df[move_col].value_counts(dropna=False))
print('\n', df[trip_col].value_counts(dropna=False))
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь

employment_types = [
    'Полная занятость', 'Частичная занятость', 'Проектная работа', 'Волонтерство', 'Стажировка'
]

if 'Занятость' in df.columns:
    for emp_type in employment_types:
        df[emp_type] = df['Занятость'].str.contains(emp_type.lower()).astype(bool)
    df.drop(['Занятость'], axis=1, inplace=True)
else:    
    print(f"Column 'Занятость' already parsed")

# Создание бинарных признаков для графика работы
schedule_types = [
    'Полный день', 'Сменный график', 'Гибкий график', 'Удаленная работа', 'Вахтовый метод'
]

if 'График' in df.columns:
    for sch_type in schedule_types:
        df[sch_type] = df['График'].str.contains(sch_type.lower()).astype(bool)
    df.drop(['График'], axis=1, inplace=True)
else:
    print(f"Column 'График' already parsed")

# df.drop(['Занятость', 'График'], axis=1, inplace=True)
```
